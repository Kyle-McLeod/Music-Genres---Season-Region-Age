```python
import pandas as pd
from pathlib import Path
```

```python
file = Path('Resources/spotify_dataset.csv')
```

```python
music_df = pd.read_csv(file, encoding="UTF-8")
music_df.head()
```

```python
music_df = music_df.sort_values("Week of Highest Charting")
music_df
```

```python
music_df = music_df[['Highest Charting Position', 'Week of Highest Charting', 'Song Name', 'Artist', 'Genre']]
music_df
```

```python
music_df = music_df.set_index('Week of Highest Charting')
music_df
```

```python
music_df = music_df["Genre"].replace({"['pop', 'poet-teen pop']": "Pop"})
music_df
```
