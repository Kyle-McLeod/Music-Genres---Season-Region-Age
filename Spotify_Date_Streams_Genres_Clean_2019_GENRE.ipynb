```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = Path('2017-2021_Spotify_Top_200.csv')
```

```python
music_df = pd.read_csv(file, encoding="UTF-8")
music_df.head()
```

```python
music_df = music_df[["Position#Track Name#Artist#Streams#Date#Genre"]]
music_df
```

```python
music_df = music_df['Position#Track Name#Artist#Streams#Date#Genre'].str.split('#', expand=True)

music_df
```

```python
music_df.columns = ["Chart Position", "Song Title", "Artist Name", "Stream Count", "Date", "Genre", "Null"]
music_df = music_df[["Chart Position", "Song Title", "Artist Name", "Stream Count", "Date", "Genre"]]
music_df.head()
```

```python
music_df[["Year", "Month", "Day"]] = music_df["Date"].str.split('-', expand=True)
music_df.head()
```

```python
music_df = music_df[["Chart Position", "Song Title", "Artist Name", "Stream Count", "Date", "Genre", "Year"]]
music_df.head()
```

```python
music_df["year"] = music_df["Year"].astype("float64")
music_df.dtypes
```

```python
music_df = music_df[["Stream Count", "Date", "Genre", "Year"]]
music_df.head()
```

```python
music_df = music_df.set_index("Year")
music_df.head()
```

```python
rows_ = ["2019"]
columns_ = ["Stream Count", "Date", "Genre"]
music_df = music_df.loc[rows_,columns_]
music_df.head()
```
