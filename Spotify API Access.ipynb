```python
import requests
client_id = '50279baf57a247eaaca7c996217ef1da'
client_secret = '43cbff499d544111806864de68d75fa5'
```

```python
auth_url = 'https://accounts.spotify.com/api/token'
```

```python
data = {
    'grant_type': 'client_credentials',
    'client_id': client_id,
    'client_secret': client_secret,
}
```

```python
auth_response = requests.post(auth_url, data=data)
```

```python
access_token = auth_response.json().get('access_token')
```

```python
base_url = 'https://api.spotify.com/v1/'
```

```python
headers = {
    'Authorization': 'Bearer {}'.format(access_token)
}
```

```python
featured_playlists_endpoint = 'browse/featured-playlists/?limit=50'
featured_playlists_url = ''.join([base_url,featured_playlists_endpoint])
```

```python
response = requests.get(featured_playlists_url,headers=headers)
```

```python
playlists = response.json().get('playlists').get('items')
playlists
```
